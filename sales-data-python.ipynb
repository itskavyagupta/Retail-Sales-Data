```python
#  1. Import necessary libraries and read the provided dataset (online_sales.csv) and check the top 5
# and random 5 samples of the dataframe.

import numpy as np
import pandas as pd

df=pd.read_csv("online_sales.csv")
df
```

```python
# Observations inferred from the info of dataframe:

# 1. There are 240007 rows present.
# 2. There are 8 columns in the given dataset namely: InvoiceNo, StockCode, Description, Quantity, InvoiceDate,
#    UnitPrice, CustomerID and Country.
# 3. Description and CustomerID columns have null value entries while the others do not.

# 4. There are 3 data types present in the given dataset, which are: float, int and object.

# 5. The shape of the dataset is: (240007, 8), i.e. 240007 rows and 8 columns.
```

```python
# 3. Check for null values and report the percentage of null values of each column.
# And drop the rows having null values in it.

df.isnull().any()
```

```python
df.isnull().sum() * 100 / len(df)
```

```python
# Comments:

# Thus, as we see from the outputs, only the Description and CustomerID columns have null values and the percentage of
# null values in them are 0.375406% and 28.009600% respectively.
```

```python
df = df.dropna()
df.isnull().any()
```

```python
# Comments:

# We have now dropped the rows having null values in them. This is verified as now all columns return 'False' when checking
# for any null values in them.
```

```python
# 4. Check the statistical summary of the dataset. (Use inbuilt functions)

df.describe()
```

```python
# 5. Drop the instances having quantity less than zero. (Note:- refer to the â€˜Quantityâ€™ feature)

pd.options.mode.chained_assignment = None

df.drop(df[df["Quantity"] < 0].index, inplace = True)

print("\n We have now dropped the rows having Quantity less than zero.")
```

```python
# 6. Check unique values of the country and report the name of the country that has the highest number of instances/rows.

df["Country"].unique()
```

```python
print("\n The 'Country' column has a  total of {} unique values.".format(len(pd.unique(df['Country']))))
```

```python
df["Country"].value_counts()[:10]
```

```python
# Comments:

# Thus, the country that has the highest number of instances is 'United Kingdom'
```

```python
# 7. Create a new column with the name as â€˜salesâ€™ having total sales. The total sales is defined as Quantity*UnitPrice.

df["sales"] = df["Quantity"] * df["UnitPrice"]
df

#We have now a column named â€˜salesâ€™ having total sales.
```

```python
# 8. Report the top 5 countries in terms of sales.
# a. Consider the size of sales.
# b. Consider the mean value of sales.

df.groupby("Country").agg({'sales':'sum'}).sort_values('sales', ascending=False)[:5]
```

```python
# Comments:

# The top 5 countries in terms of sales are: United Kingdom, Netherlands, Germany, EIRE and France.
```

```python
# Reporting the size of sales:

df.groupby("Country").agg({'sales':'size'}).sort_values('sales', ascending=False)
```

```python
# Reporting the mean value of sales:

df.groupby("Country").agg({'sales':'mean'}).sort_values('sales', ascending=False)
```

```python
# 9. Report the top 5 products which bring the highest sales. Use StockCode for product information.

df.groupby(['StockCode', 'Description'])['sales'].sum().to_frame('sales').sort_values('sales', ascending=False)[:5]
```

```python
# Comments:

# Thus, the top 5 products which bring the highest sales are: REGENCY CAKESTAND 3 TIER, MEDIUM CERAMIC TOP STORAGE JAR,
# WHITE HANGING HEART T-LIGHT HOLDER, PARTY BUNTING and POSTAGE.
```

```python
# 10. Convert the â€˜InvoiceDateâ€™ into a date format and report the month on which the maximum sales occur?

df.info()
print("\n")

df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], format='%m/%d/%y %H:%M')
df.info()

print("\n We have now converted the 'InvoiceDate' column into a date format which can be seen as\
 the data type of the 'InvoiceDate' column has changed from object to 'datetime'.")
```

```python
df_new=df

df_new['InvoiceDate']=pd.to_datetime(df_new['InvoiceDate'])
df_new.groupby(df['InvoiceDate'].dt.month).agg({'sales':'max'})['sales'].nlargest()
```

```python
# Comments:

# As inferred from the output, the month which has the maximum sales is May.
```
